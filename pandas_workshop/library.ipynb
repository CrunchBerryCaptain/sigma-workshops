```python
import pandas as pd
import altair as alt


alt.data_transformers.enable("vegafusion")

library = pd.read_csv("library_users_clean.csv", index_col=0)

library.head(1)
```

```python
alt.Chart(library).mark_bar().encode(
    x=alt.X("Patron Type Definition:O", sort="-y", title='Patron Type'),
    y=alt.Y("count():Q", title='Total Patrons'),
    color=alt.Color("Patron Type Definition:O", scale=alt.Scale(
        scheme='category20b')),  # Set a color scheme
    # Add tooltip for interactivity
    tooltip=['Patron Type Definition', 'count():Q']
).properties(
    width=800,
    height=400,
    title=alt.TitleParams(
        text='Library Patrons by Type',
        subtitle='Total count of patrons based on type',
        align='center',
        fontSize=20
    )
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=24,
    anchor='middle',
    color='navy'
)
```
